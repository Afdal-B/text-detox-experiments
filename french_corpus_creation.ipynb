```python
# deepl
auth_key = "xxxx"
deepl_client = deepl.DeepLClient(auth_key)

def translate_with_deepl(text):
    try:
        return deepl_client.translate_text(text,target_lang='fr').text
    except Exception as e:
        return f"[DeepL error] {e}"
```

```python
!pip install googletrans==4.0.0-rc1
from googletrans import Translator

translator = Translator()

def translate_with_google(text):
    try:
        return translator.translate(text, dest='fr').text
    except Exception as e:
        return f"[Google Translate error] {e}"
```

```python
# Login using e.g. `huggingface-cli login` to access this dataset
df = pd.read_csv("hf://datasets/s-nlp/paradetox/train.tsv", sep="\t")
```

```python
french_df = pd.DataFrame()
```

```python
sample = df.sample(6000)
```

```python
french_df["fr_toxic_comment"] = sample["en_toxic_comment"].progress_apply(translate_with_google)
french_df["fr_neutral_comment"] = sample["en_neutral_comment"].progress_apply(translate_with_google)
```

```python
french_df.to_csv("french_paradetox_6000.csv",sep="\t")
```
