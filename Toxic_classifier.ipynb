```python
from datasets import load_dataset, Dataset
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

dataset = load_dataset("textdetox/multilingual_toxicity_dataset")

train_texts, train_labels, train_langs = [], [], []
test_texts, test_labels, test_langs = [], [], []

for lang in dataset.keys():
    texts = [sample["text"] for sample in dataset[lang]]
    labels = [sample["toxic"] for sample in dataset[lang]]
    langs = [lang] * len(texts)

    X_train, X_test, y_train, y_test = train_test_split(
        texts, labels, test_size=0.5, stratify=labels, random_state=42
    )

    train_texts.extend(X_train)
    train_labels.extend(y_train)
    train_langs.extend(langs[:len(X_train)])

    test_texts.extend(X_test)
    test_labels.extend(y_test)
    test_langs.extend(langs[:len(X_test)])

train_df = pd.DataFrame({
    "text": train_texts,
    "label": train_labels,
    "lang": train_langs
})

test_df = pd.DataFrame({
    "text": test_texts,
    "label": test_labels,
    "lang": test_langs
})

train_df.to_csv("/content/drive/MyDrive/tsv_files/multilingual_toxicity_train.csv", index=False)
test_df.to_csv("/content/drive/MyDrive/tsv_files/multilingual_toxicity_test.csv", index=False)

```

```python
train_df = pd.read_csv("multilingual_toxicity_train.csv")
test_df = pd.read_csv("multilingual_toxicity_test.csv")

train_dataset = Dataset.from_pandas(train_df).remove_columns(["lang"])
test_dataset = Dataset.from_pandas(test_df).remove_columns(["lang"])

tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")

def tokenize(example):
    return tokenizer(example["text"], truncation=True, padding="max_length")

tokenized_train = train_dataset.map(tokenize, batched=True)
tokenized_test = test_dataset.map(tokenize, batched=True)

tokenized_train = tokenized_train.rename_column("label", "labels")
tokenized_test = tokenized_test.rename_column("label", "labels")

tokenized_train.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
tokenized_test.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])

model_name = "xlm-roberta-base"
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds)
    }

training_args = TrainingArguments(
    output_dir="./toxicity_classifier",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=16,
    logging_steps=100,
    eval_steps=100,
    num_train_epochs=3,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    eval_strategy="epoch",
    learning_rate=2e-4,
    report_to = "none"

)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()
```
