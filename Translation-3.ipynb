```python
async def translate_text(text, target_language):
    # Translation
    try:
        translated = await translator.translate(text, dest=target_language)
        return translated
    except Exception as e:
        print(f"Translation error: {e}")
        return text
```

```python
import asyncio
import pandas as pd
from tqdm import tqdm

async def translate_row(toxic, neutral, target_language):
    toxic_task = translate_text(toxic, target_language)
    neutral_task = translate_text(neutral, target_language)
    toxic_translated, neutral_translated = await asyncio.gather(toxic_task, neutral_task)
    return toxic_translated, neutral_translated

async def translate_dataset(dataset, target_language):
    toxic_texts = dataset.iloc[:, 0].tolist()
    neutral_texts = dataset.iloc[:, 1].tolist()

    toxic_results = []
    neutral_results = []

    print("Démarrage de la traduction...")

    for i in tqdm(range(len(dataset)), desc=f"Translating into {target_language}"):
        toxic_trad, neutral_trad = await translate_row(toxic_texts[i], neutral_texts[i], target_language)
        toxic_results.append(toxic_trad)
        neutral_results.append(neutral_trad)

    translated_df = pd.DataFrame({
        "toxic_sentences": toxic_results,
        "neutral_sentences": neutral_results
    })

    return translated_df
```

```python
#langs = ["he","ja","de","es","zh","ar","hi","am"]
langs =["fr"]
translator = Translator()

for lang in langs:
  translated_df=await translate_dataset(df,target_language=lang)
  translated_df['toxic_text'] = translated_df['toxic_sentences'].apply(lambda x: x.text if hasattr(x, 'text') else x)
  translated_df['neutral_text'] = translated_df['neutral_sentences'].apply(lambda x: x.text if hasattr(x, 'text') else x)
  translated_df = translated_df[['toxic_text', 'neutral_text']]
  translated_df.to_csv(f'/content/drive/MyDrive/training_dataset/translated/translated_data_es_to_{lang}.csv', index=False)
```
